```python
from pylab import*
from scipy.integrate import*
%matplotlib inline
```

```python
def osc(y_v,t,omega2=1.,mu=0):
    #y_v[0] es y, y_v[1] es v 
    dy = y_v[1]
    dv = -omega2*y_v[0]-mu*y_v[1]
    return array([dy,dv])
```

```python
t = linspace(0,100,500)
inicio = [0.,1.]
sol=odeint(osc,inicio,t,args=(1.,0.1))
xlim([-1,1]),ylim([-1,1])
plot(sol[:,0],sol[:,1])
```

```python
from JSAnimation import IPython_display
from matplotlib import animation
```

```python
figura = figure()
ejes = axes(xlim=(-1,1),ylim=(-1,1))

trayectoria, = ejes.plot([],[],lw=2)
punto, = ejes.plot([],[],'o')

def init():                     # esta funcion la vamos a usar para reiniciar la animacion
    trayectoria.set_data([],[])
    punto.set_data([],[])
    return trayectoria,punto,

def el_mov(i):                #la funcion que queremos animar
    y,v =sol[:,0],sol[:,1]
    trayectoria.set_data(y[:i+1],v[:i+1])#
    punto.set_data(y[i],v[i])
    return trayectoria,punto

animation.FuncAnimation(figura,el_mov,init_func=init,frames=20) #en esta parte hacemos la animacion, lo que hace es que grafica la solucion,borra,grafica y borra hasta llegar a 20 o al numero de frames
```
